**8.Ensemble Learning using Stacking**

```python
from sklearn.ensemble import StackingClassifier

base_models = [('SVM',SVC()),('Decision Tree',DecisionTreeClassifier()),('Logistic Regerssion',LogisticRegression()) , ('KNN',KNeighborsClassifier(n_neighbors=3))]
stacking = StackingClassifier(
    estimators = base_models ,
    final_estimator = SVC(),
    cv = 5
)

stacking.fit(X_train , y_train)
```

```python
df = pd.DataFrame(data={'Model': ['Stacking','DecisionTree', 'KNN', 'SVC', 'RF','LR'], 'Score':
                        [stacking.score(X_test,y_test),dectree.score(X_test, y_test),knn.score(X_test,y_test),svc.score(X_test,y_test),
                         ranfor.score(X_test,y_test),logreg.score(X_test,y_test)]})

plt.figure(figsize=(12, 8))
sns.barplot(x="Model", y="Score", data=df, palette="magma")
plt.title('Performance analysis of different classifiers')
plt.show()
```
